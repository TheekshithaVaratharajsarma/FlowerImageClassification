## Comparison of the 4 models

```python
import matplotlib.pyplot as plt

test_loss = [1.238970398902893, 0.7693274617195129, 0.8000471591949463, 0.8086612820625305]
test_accuracy = [0.7400000095367432, 0.7309091091156006, 0.7545454502105713, 0.7685185074806213]
variations = ['Original', 'Variation 1', 'Variation 2', 'Variation 3']

# Plotting Loss
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(variations, test_loss, marker='o', linestyle='-', color='b')
plt.title('Test Loss')
plt.xlabel('Variations')
plt.ylabel('Loss')

# Plotting Accuracy
plt.subplot(1, 2, 2)
plt.plot(variations, test_accuracy, marker='o', linestyle='-', color='r')
plt.title('Test Accuracy')
plt.xlabel('Variations')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame({'Variation': variations, 'Test Accuracy': test_accuracy, 'Test Loss': test_loss})
df
```

## Original Model

```python
# Reading training history text file
with open('training_history.txt', 'r') as file:
    history = eval(file.read())

# Plotting Loss and Validation Loss
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history['loss'], label='Training Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.title('Training History - Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plotting Accuracy and Validation Accuracy
plt.subplot(1, 2, 2)
plt.plot(history['accuracy'], label='Training Accuracy')
plt.plot(history['val_accuracy'], label='Validation Accuracy')
plt.title('Training History - Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```
